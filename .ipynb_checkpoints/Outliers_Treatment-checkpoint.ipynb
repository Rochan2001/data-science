# Removing Outliers

To remove outliers:
    1. np.log()  --> Natural log element wise
    2. np.sqrt() --> Square-Root 
    3. np.cbrt() --> Cube-Root

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('loan_new.csv')
```

```python
data['Annual Income'].min()
```

```python
data['Annual Income'].max()
```

```python
data['Annual Income'].mean()
```

```python
plt.hist(data['Annual Income'])
plt.show()
```

```python
# Removing Outliers
data_log = np.log(data['Annual Income'])
```

```python
data_log.plot.hist(alpha = 1,color = 'cyan',edgecolor = 'black')
plt.show()
```

```python
data_sqrt = np.sqrt(data['Annual Income'])
```

```python
data_sqrt.plot.hist(alpha = 0.8,color = 'yellow',edgecolor = 'black')
plt.show()
```

```python
data_cbrt = np.cbrt(data['Annual Income'])
```

```python
data_cbrt.plot.hist(alpha = 0.8,color = 'magenta',edgecolor = 'black')
plt.show()
```

```python
per = data['Annual Income']
```

```python
per1 = np.percentile(per,0.25)
```

```python
per.plot.hist(alpha = 1)
plt.plot(per1)
plt.show()
```

# Normalisation

It is the process of recognizing data in a dataset so that it meets two basic requirements:
1. There is no REDUNDANCY of sata (i.e. all data is stored in only one place).
2. Data Dependencies are logical (i.e. all related data items are stored together).

```python
import pandas as pd
```

```python
data = pd.read_csv('Documents/Data science/')
```

```python
tab = pd.crosstab(index = data['Term'],columns = "count",normalize = True)
```

```python
tab
data.head()
```

```python
tab2 = pd.crosstab(index = data['Loan Status'],columns = "count",normalize = True)
tab2
```

```python
tab3 = pd.crosstab(index = data['Purpose'],columns = "count",normalize = True)
tab3
```
